<h3>News article crawler</h3>
This code pull articles from https://cnn.com and outputs the first 50 with a text lenght over 1000 characters to a .json file including the title, publishing date, the article text, and the url to the article.

```python
import newspaper
from newspaper import Article
import json, io, newspaper

cnn_paper = newspaper.build('https://www.cnn.com/', memoize_articles=False)

data = {}
i = 0

print(len(cnn_paper.articles))
#if len(yahoo_paper) < 50 : error else run this block

for article in cnn_paper.articles:
    try :     
        article.download()
        article.parse()
        if len(article.text) > 1000 :
            #"""
            #console debugging
            i += 1
            print('Article # ', i)
            print(article.title)
            print(article.publish_date)
            print('Article lenght: ', len(article.text))
            print(article.url)
            print("\n\n\n")
            #"""

            data[i] = [article.title, article.publish_date, article.text, article.url]

            if i == 50:
                break
    except :
        pass
    
        
with open('data.json', 'w') as outfile:
    json.dump(data, outfile, indent=4, sort_keys=True, default=str)    
```
