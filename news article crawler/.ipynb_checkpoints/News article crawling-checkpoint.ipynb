```python
import newspaper
from newspaper import Article
import json, io, newspaper

cnn_paper = newspaper.build('https://www.cnn.com/')

```

```python
data = {}
i = 0

print(len(cnn_paper.articles))
#if len(yahoo_paper) < 50 : error else run this block

for article in cnn_paper.articles:
    try :     
        article.download()
        i += 1
        article.parse()

        #"""
        #console debugging
        print('Article # ', i)
        print(article.title)
        print(article.publish_date)
        print(article.text)
        print(article.url)
        print("\n\n\n")
        #"""

        data[i] = [article.title, article.publish_date, article.text, article.url]

        if i == 50:
            break
    except :
        pass
    
        
with open('data.json', 'w') as outfile:
    json.dump(data, outfile, indent=4, sort_keys=True, default=str)    
```
